```python
 # Import the dependencies
import pandas as pd
import numpy as np
import datetime

# ORM Stuff
from sqlalchemy import create_engine, inspect, text
```

```python
bees_df = pd.read_csv("save_the_bees.csv")
bees_df.head()
```

```python
bees_df.info()
```

```python
states_df = pd.read_csv("US_GeoCode.csv")
states_df.head()
```

```python
states_df.info()
```

```python
# Merge the dataframes on the state name
df = bees_df.merge(states_df[['Name', 'latitude', 'longitude']], left_on='state', right_on='Name', how='left')
df.head()
```

```python
df.info()
```

```python
# Drop the duplicate 'Name' column since 'state' already contains the state name
df.drop(columns=['Name'], inplace=True)
df.head()
```

```python
# Check for missing lat/lon
missing_geo = df[df['latitude'].isna()]
if not missing_geo.empty:
    print("⚠️ Warning: Some states are missing latitude & longitude.")
    print(missing_geo[['state', 'state_code']])
```

```python
# "Other" possibly combines the data from the missing states AK, DE, DC, NV, NH
```

```python
# Save the cleaned dataset
df.to_csv("cleaned_bees_data.csv", index=False)
```

```python
# WRITE TO DATABASE
engine = create_engine("sqlite:///bees.sqlite")
```

```python
# Random sample for SPEED
df2 = df.sample(500)
df2.head()
```

```python
 # Write to SQL
df.to_sql(name="bees", con=engine, index=False, if_exists="append", method="multi")
```

```python
# Create the inspector and connect it to the engine
inspector = inspect(engine)

# Collect the names of tables within the database
tables = inspector.get_table_names()

# Using the inspector to print the column names within the 'cleaned_bees_data' table and its types
for table in tables:
    print(table)
    print("--------")
    columns = inspector.get_columns(table)
    for column in columns:
        print(column["name"], column["type"])

    print()
```

```python
engine.dispose()
```
